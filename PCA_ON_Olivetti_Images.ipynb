```
#import files
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_olivetti_faces
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
```

```
images=fetch_olivetti_faces()['images']

images.shape
```

```
fig1=plt.figure(figsize=(12,12))
for i in range(0,50):
  ax=fig1.add_subplot(8,8,i+1)
  ax.imshow(images[i],cmap=plt.cm.bone)
plt.show()
```

```
x=fetch_olivetti_faces()['data']
y=fetch_olivetti_faces()['target']
```

```
pca=PCA(123)
x=pca.fit_transform(x)
pca.explained_variance_ratio_
```

```
k_factor=90
for i in range(len(pca.explained_variance_ratio_)):
  sum=pca.explained_variance_ratio_[:i].sum()
  if sum*100>k_factor:
    break
print(i)
```

```
x=pca.inverse_transform(x)
x=x.reshape(400,64,64)
fig1=plt.figure(figsize=(12,12))
for i in range(0,50):
  ax=fig1.add_subplot(8,8,i+1)
  ax.imshow(x[i],cmap=plt.cm.bone)
plt.show()
```

```
eigen_faces=pca.components_
eigen_faces=eigen_faces.reshape(-1,64,64)
fig1=plt.figure(figsize=(12,12))
for i in range(0,50):
  ax=fig1.add_subplot(8,8,i+1)
  ax.imshow(eigen_faces[i],cmap=plt.cm.bone)
plt.show()
```
